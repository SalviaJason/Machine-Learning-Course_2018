```python
%matplotlib inline
import numpy as np
import random
import matplotlib.pyplot as plt
```

#  Random Data Generator

a. Univariate gaussian data generator

```python
def UGDG(m,s):
    U = random.uniform(0, 1)
    V = random.uniform(0, 1)
    z = np.sqrt(-2*np.log(U))*np.cos(2*np.pi*V)
    X = m + s*z
    return X
```

b. Polynomial basis linear model data generator

```python
def PBLM(n, a, w):
    x = random.uniform(-1, 1)
    e = np.random.normal(0, a, 1)
    y = 0
    for i in range(n):
        y = w[0][i]*pow(x,i) + y
    y = y + e
    return x,y
```

# Sequential Estimator

```python
n = 1
m0 = m = 3.0
s0 = s = 5.0
print("Data point source function: N(",m, ", ",s,")")
print(" ")
x = UGDG(m0,s0)
while ((np.square(m-m0)+np.square(s-s0)) >= 0):
    n += 1
    x = UGDG(m0,s0)
    print("Add data point: ",x)
    mnew = m  + (x-m)/n
    s2new = np.square(s) + (np.square(x-m)/n) - (np.square(s)/(n-1))
    snew = np.sqrt(s2new)
    m = mnew
    s = snew
    print("Mean =",m,"   Variance =" ,s)
    print("")
    if (np.square(m-m0)+np.square(s-s0)) < 0.01 :
        break
    
```

# Baysian Linear regression

```python
b = 1
n = 4
a = 1
w = np.array([[1, 2, 3, 4]])
I = np.identity(n)
s = b*I
m = np.zeros([n,1])

for i in range(1500):
    (x,y)= PBLM(n, a, w)
    print("Add data point (",x,",", y[0],"):")
    print("")
    
    X = []
    for i in range(n):
        X.append(pow(x,i))
    X = (np.array(X)).reshape(1,n)
    outputx = np.dot(X,m)[0,0]
    outputy = a + np.dot(np.dot(X,np.linalg.inv(s)),X.T)[0,0]
    
    snew = a*np.dot((X.T), X) + s
    mnew = np.dot(np.linalg.inv(snew), a*(X.T)*y + np.dot(s, m))
    s = snew
    m = mnew 
    print("Postirior mean:", "\n",m)
    print("")
    print("Posterior variance:","\n", np.linalg.inv(s))
    print("")
    print("Predictive distribution ~ N(",outputx, ",",outputy,")")
    print("---------------------------------------------------------------")

xline = np.linspace(-2, 2, 100)
yline = np.zeros(shape=(100,))
yl = yline*0
for i in range(n):
    yline = m[i]*pow(xline,i) + yline
    yl = np.linalg.inv(s)[i][i]*pow(xline,i) + yl

plt.plot(xline,yline,'k-') # mean
plt.plot(xline,yline+ outputy,'r-')
plt.plot(xline,yline- outputy,'r-')
plt.ylim(-20, 25)
plt.show()
```
